```python
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import numpy as np
from matplotlib import pyplot as plt
import pickle 
from sklearn.tree import plot_tree
```

```python
iris = load_iris()
iris
```

```python
for i, j in zip(range(4), iris.feature_names):
    print(j)
    print("min = ", min(iris.data[:,i]))
    print("max = ", max(iris.data[:,i]))
```

```python

_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
plt.savefig("scat.jpg", dpi=900)
```

```python
x = iris.data
y = iris.target
```

```python
test_ids = np.random.permutation(len(x))
```

```python
x_train = x[test_ids[:-20]]
x_test = x[test_ids[-20:]]
```

```python
y_train = y[test_ids[:-20]]
y_test = y[test_ids[-20:]]
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
predicted_values = model.predict(x_test)
```

```python
predicted_values
```

```python
acc_score = accuracy_score(predicted_values, y_test)*100
```

```python
acc_score
```
